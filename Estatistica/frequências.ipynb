```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 
```

```python
tb_people_contry = pd.read_csv(r"C:\Users\luiza\OneDrive\Área de Trabalho\av neuropsicologica\mba_usp\Arquivos\Exercicio - Estatística  - Tabela Pessoas-Local .csv")
tb_people_contry
```

**Frequência Absoluta**

*Contagem de ocorrências em categoria*

```python
country_counts = tb_people_contry["País de Origem"].value_counts().sort_values(ascending=True)
country_counts
```

```python
labels = country_counts.index
sizes = country_counts.values
num_categories = len(labels)
colors = plt.cm.viridis(np.linspace(0, 1, num_categories))
fig, ax = plt.subplots()
bar_height = 0.5
bars = ax.barh(labels, sizes, height=bar_height, color=colors)
ax.set_title('Quantidade de Pessoas por País')
ax.set_xlabel('Quantidade de Pessoas')
ax.set_ylabel('País')
ax.margins(x=0.1) 
for bar in bars:
    width = bar.get_width()
    ax.annotate(f'{width}',
                xy=(width, bar.get_y() + bar.get_height() / 2),
                xytext=(3, 0),  
                textcoords="offset points",
                ha='left', va='center')
fig.tight_layout()
plt.show()
```

**Frequência Relativa**

*Percentual de cada categoria em relação ao total de observações*

```python

total_pessoas = country_counts.sum()
frequencia_relativa = country_counts / total_pessoas
labels = country_counts.index
sizes = country_counts.values
num_categories = len(labels)
colors = plt.cm.viridis(np.linspace(0, 1, num_categories))
fig, ax = plt.subplots()
bar_height = 0.5
bars = ax.barh(labels, sizes, height=bar_height, color=colors)
ax.set_title('Quantidade de Pessoas por País')
ax.set_xlabel('Quantidade de Pessoas')
ax.set_ylabel('País')
ax.margins(x=0.1) 
for bar, freq in zip(bars, frequencia_relativa):
    width = bar.get_width()
    ax.annotate(f'{width} ({freq:.2%})',  
                xy=(width, bar.get_y() + bar.get_height() / 2),
                xytext=(3, 0), 
                textcoords="offset points",
                ha='left', va='center')
fig.tight_layout()
plt.show()
```

**Frequência Acumulada** 

*Soma de frequência absoluta a cada nova categoria*

```python
country_counts = tb_people_contry["País de Origem"].value_counts().sort_values(ascending=False)
frequencia_acumulada = country_counts.cumsum()
labels = country_counts.index
sizes = country_counts.values
num_categories = len(labels)
colors = plt.cm.viridis(np.linspace(0, 1, num_categories))
fig, ax = plt.subplots()
bar_height = 0.5
bars = ax.barh(labels, sizes, height=bar_height, color=colors)
ax.set_title('Quantidade de Pessoas por País')
ax.set_xlabel('Quantidade de Pessoas')
ax.set_ylabel('País')
ax.margins(x=0.1)
for bar, acumulado in zip(bars, frequencia_acumulada):
    width = bar.get_width()
    ax.annotate(f'{width} / {acumulado}', 
                xy=(width, bar.get_y() + bar.get_height() / 2),
                xytext=(3, 0),  
                textcoords="offset points",
                ha='left', va='center')
fig.tight_layout()
plt.show()
```
