```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import statistics
```

```python
tabela_precos = pd.read_csv(r"C:\Users\luiza\OneDrive\Área de Trabalho\av neuropsicologica\mba_usp\Arquivos\Exercicio - Estatística  - Tabela Preços - Produtos .csv")

tabela_precos["Preço ($)"] = tabela_precos["Preço ($)"].str.replace(",", "").astype(int)
tabela_precos["Preço ($)"] = tabela_precos["Preço ($)"]/100
tabela_precos
```

```python

plt.hist(tabela_precos["Preço ($)"], bins=20, edgecolor='black', color='skyblue')  # 'bins' define o número de caixas no histograma

plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma')
plt.show()
```

```python
ordenado = tabela_precos['Preço ($)'].sort_values(ascending = True)
print(ordenado)

```

**Média** 

É a soma dos valores contidos, divido pela quantidade total de obsevações 

A média é muito utilizada como ponto de referência das análises (sendo uma medida de tendência central)

A atenção que devemos ter é se temos observações que tem diferentes amplitudes pode a média pode não ser a medida mais adequada como ponto de referência 

```python
media = statistics.mean(ordenado)

print("A média é:", media)
```

```python
import matplotlib.pyplot as plt
num_bins = 20
alpha = 0.7
color = 'skyblue'
edgecolor = 'black'
label = 'Dados'
plt.hist(tabela_precos['Preço ($)'], bins=num_bins, alpha=alpha, color=color, edgecolor=edgecolor, label=label)
media = sum(tabela_precos['Preço ($)']) / len(tabela_precos['Preço ($)'])
plt.axvline(x=media, color='red', linestyle='--', label='Média')
plt.legend()
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma')
plt.show()
```

**Mediana** 

É o elemento da distribuição de valores, sempre considerando que os valores estão ordenados de forma crescente 

```python
mediana = statistics.median(ordenado)

print("A mediana é:", mediana)
```

```python
import matplotlib.pyplot as plt
num_bins = 20
alpha = 0.7
color = 'skyblue'
edgecolor = 'black'
label = 'Dados'
plt.hist(tabela_precos['Preço ($)'], bins=num_bins, alpha=alpha, color=color, edgecolor=edgecolor, label=label)
media = sum(tabela_precos['Preço ($)']) / len(tabela_precos['Preço ($)'])
plt.axvline(x=mediana, color='purple', linestyle='--', label='Média')
plt.legend()
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma')
plt.show()
```

**Moda** 

O valor que aparece com maior frequência


```python
moda = statistics.mode(ordenado)

print("A moda é:", moda)
```
